```python
questions_from_schema = set(schema_df['qname'])   #питання зі схеми
columns_in_df = set(df.columns)                   #колонки з основної таблиці

#спільні назви (інтерсекція)
common_columns = list(questions_from_schema.intersection(columns_in_df))
print(len(common_columns))
```

__3. Статистичний аналіз досвіду респондентів:__

```python
#показники для колонки WorkExp
mean_val = df['WorkExp'].mean()
median_val = df['WorkExp'].median()
mode_val = df['WorkExp'].mode()[0] # Мода повертає список, тому обрано перший елемент


print(f"Середній досвід роботи: {mean_val:.2f} років")
print(f"Медіанний досвід роботи: {median_val} років")
print(f"Найпопулярніший досвід роботи(мода): {mode_val} років")
```

__5. Визначення популярності Python:__

```python
[col for col in df.columns if 'Language' in col]
```

```python
df['LanguageHaveWorkedWith'].head()    
```

```python
python_users = df['LanguageHaveWorkedWith'].apply(lambda x: 'Python' in str(x))

python_count = python_users.sum()   # кількість тих, хто знає Python
python_count
```

```python
percent = (python_count / df['LanguageHaveWorkedWith'].count()) * 100

print(f"Відсоток респондентів, які програмують на Python: {round(percent, 1)}%")
```

__7. Географічний аналіз компенсації Python-розробників:__

```python
python_df = df[python_users].copy()   #всі рядки з df, де python_users True
```

```python
#середнє та медіана для зарплати
salary_analysis = python_df.groupby('Country')['ConvertedCompYearly'].agg(['mean', 'median'])
salary_analysis.columns = ['Mean salary', 'Median salary']
salary_analysis
```

__10. Аналіз індустрій серед високооплачуваних віддалених працівників:__

```python
salary_limit = df['ConvertedCompYearly'].quantile(0.75)
salary_limit 
```

```python
rich_remote = (df['ConvertedCompYearly'] > salary_limit) & (df['RemoteWork'] == 'Remote')
top_rich_remote = df[rich_remote]

result = top_rich_remote['Industry'].value_counts()
result
```
